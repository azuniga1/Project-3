````python
starting_lineups.drop(columns=['Box Score'], inplace=True)
````

````python
starting_lineups.Team.drop_duplicates()
````

````python
starting_lineup_encoder = encoded_box_scores[['Team','date','game_id','Location','Outcome']].copy()
print(len(starting_lineup_encoder))
starting_lineup_encoder.drop_duplicates(inplace=True)
print(len(starting_lineup_encoder))
````

This makes sense since around 10 players per team are active in any given game.

````python
starting_lineup_encoder.sample(50)
````

````python
team_to_abbr_dict = {'ATLANTA_HAWKS': 'ATL',
 'BOSTON_CELTICS': 'BOS',
 'BROOKLYN_NETS': 'BRK',
 'CHARLOTTE_BOBCATS': 'CHO',
 'CHARLOTTE_HORNETS': 'CHO',
 'CHICAGO_BULLS': 'CHI',
 'CLEVELAND_CAVALIERS': 'CLE',
 'DALLAS_MAVERICKS': 'DAL',
 'DENVER_NUGGETS': 'DEN',
 'DETROIT_PISTONS': 'DET',
 'GOLDEN_STATE_WARRIORS': 'GSW',
 'HOUSTON_ROCKETS': 'HOU',
 'INDIANA_PACERS': 'IND',
 'LOS_ANGELES_CLIPPERS': 'LAC',
 'LOS_ANGELES_LAKERS': 'LAL',
 'MEMPHIS_GRIZZLIES': 'MEM',
 'MIAMI_HEAT': 'MIA',
 'MILWAUKEE_BUCKS': 'MIL',
 'MINNESOTA_TIMBERWOLVES': 'MIN',
 'NEW_ORLEANS_HORNETS': 'NOP',
 'NEW_ORLEANS_PELICANS': 'NOP',
 'NEW_YORK_KNICKS': 'NYK',
 'OKLAHOMA_CITY_THUNDER': 'OKC',
 'ORLANDO_MAGIC': 'ORL',
 'PHILADELPHIA_76ERS': 'PHI',
 'PHOENIX_SUNS': 'PHO',
 'PORTLAND_TRAIL_BLAZERS': 'POR',
 'SACRAMENTO_KINGS': 'SAC',
 'SAN_ANTONIO_SPURS': 'SAS',
 'TORONTO_RAPTORS': 'TOR',
 'UTAH_JAZZ': 'UTA',
 'WASHINGTON_WIZARDS': 'WAS'}
````

````python
len(starting_lineup_encoder[['Team']].drop_duplicates())
````

````python
starting_lineup_encoder.head()
````

````python
starting_lineup_encoder.Team = starting_lineup_encoder.Team.apply(lambda x: team_to_abbr_dict[x])
starting_lineup_encoder.head()
````

````python
starting_lineups.rename(columns = {'Game Date':'date'}, inplace=True)
````

````python
starting_lineups.dtypes
````

````python
encoded_lineup = pd.merge(starting_lineups,starting_lineup_encoder,on=['date','Team'],how='left')
encoded_lineup.sample(20)
````
