```python
from basketball_reference_web_scraper import client
from basketball_reference_web_scraper.data import Team
import pandas as pd
import numpy as np
```

```python
import datetime
from datetime import date
start = datetime.datetime(2013, 1, 1)
end = date.today()
mydates = pd.date_range(start, end)#.tolist()
```

```python
def allscores(day, month, year):
    df = pd.DataFrame(client.player_box_scores(day=day, month=month, year=year))
    return df
```

```python
def all_data(y, m, d):
    start = datetime.datetime(y, m, d)
    end = date.today()
    mydates = pd.date_range(start, end)#.tolist()

    for x in mydates:
        day = x.day
        month = x.month
        year = x.year
        allscores(day, month, year)
```

```python
df = []
for x in mydates:
        day = x.day
        month = x.month
        year = x.year
        day_plays = allscores(day, month, year)
        day_plays['date'] = x
        df.append(day_plays)
```

```python
len(df)
```

```python
joined = pd.DataFrame()

for x in range(len(df)):
    joined = pd.DataFrame.append(joined, df[x], ignore_index=True, sort=False)
```

```python
joined
```

```python
joined.to_csv('2013-2020.csv')
```
