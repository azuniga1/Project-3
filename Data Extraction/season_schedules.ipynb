## Get Season Schedules

```python
from basketball_reference_web_scraper import client
import pandas as pd
```

```python
def season_schedule(year):
    df = pd.DataFrame(client.season_schedule(season_end_year=year))
    return df
```

```python
last_8_seasons = pd.concat([season_schedule(year) for year in range(2013,2021)],
                          axis = 'rows',
                          join = 'inner')
```
